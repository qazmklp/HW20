```python
# Clustering/Dimensionality Reduction
from sklearn.manifold import TSNE
from sklearn.cluster import DBSCAN
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

# DataScience
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Prepare the Data

df = pd.read_csv("myopia.csv")
df.head()
```

```python
df.drop(columns=["MYOPIC"], inplace=True)
df.head()
```

```python
df.corr()
```

```python
X = TSNE(learning_rate="auto").fit_transform(df)

plt.figure(figsize=(12,8))

plt.scatter(X[:,0], X[:,1])
plt.title("TSNE Plot")
plt.show()
```

```python
list(zip(range(len(df.columns)), pca.explained_variance_ratio_))
```

```python
pca = PCA(n_components=4)
pca.fit(df)

components = pd.DataFrame(pca.transform(df), columns=["PCA1", "PCA2", "PCA3", "PCA4"])

components.head()
```

```python
X = TSNE(learning_rate="auto").fit_transform(components)

plt.figure(figsize=(12,8))

plt.scatter(X[:,0], X[:,1])
plt.title("TSNE Plot")
plt.show()
```

```python
inertia = []
k_range = list(range(2, 20))

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df)
    
    inertia.append(kmeans.inertia_)
    
# make plot
plt.figure(figsize=(10,6))

plt.plot(k_range, inertia)
plt.xlabel("Number of Clusters")
plt.ylabel("Inertia")
plt.title("Elbow Plot")

plt.show()
```

```python
inertia = []
k_range = list(range(2, 20))

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(components)
    
    inertia.append(kmeans.inertia_)
    
# make plot
plt.figure(figsize=(10,6))

plt.plot(k_range, inertia)
plt.xlabel("Number of Clusters")
plt.ylabel("Inertia")
plt.title("Elbow Plot")

plt.show()
```

```python
kmeans = KMeans(n_clusters=8, random_state=42)
kmeans.fit(df)
preds = kmeans.predict(df)

plt.figure(figsize=(12,8))

plt.scatter(X[:,0], X[:,1], c=preds)
plt.title("TSNE Plot")
plt.show()
```

```python
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(df)
preds = kmeans.predict(df)

plt.figure(figsize=(12,8))

plt.scatter(X[:,0], X[:,1], c=preds)
plt.title("TSNE Plot")
plt.show()
```

```python
mergings = linkage(df, method="ward")

plt.figure(figsize=(20,5))

dendrogram(mergings,
          leaf_rotation=90,
          leaf_font_size=5)

plt.show()
```

```python
cluster = AgglomerativeClustering(
    n_clusters=5,
    affinity="euclidean",
    linkage="ward"
)

preds = cluster.fit_predict(df)
plt.figure(figsize=(12,8))

plt.scatter(X[:,0], X[:,1], c=preds)
plt.title("TSNE Plot")
plt.show()
```
